```python
#Covert Normal Distribution to Standrd Normal Distribution using inline for loop method Traditional for loop Method
def stdNDgraph(dataset):  
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    
    #convert dataset to list
    values=[]
    for i in dataset:
        values.append(i)
        
    #Calulate z_score
    z_score=[]
    for j in values:
        z=(j-mean)/std
        z_score.append(z)
        
    #Plot distribution
    sns.displot(z_score,kde=True)

    #return Average of z-score
    return sum(z_score)/ len(z_score)
```

```python
stdNDgraph(dataset["ssc_p"])
```

```python
#Covert Normal Distribution to Standrd Normal Distribution using inline for loop method
def stdNDgraph(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    
    #convert dataset to list
    values=[i for i in dataset]
    
    #Calulate z_score
    z_score=[((j-mean)/std) for j in values]
    #Plot distribution
    sns.displot(z_score,kde=True)
    
    #return Average of z-score
    return sum(z_score)/len(z_score)
```
