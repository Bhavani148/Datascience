```python
# Step 1: Separate quantitative and qualitative
quan, qual = Univariate.quanQual(dataset)
print("Quantitative:", quan)
print("Qualitative:", qual)
```

```python
# Step 2: Descriptive statistics
descriptive = Univariate.descriptive_stats(dataset, quan)
print(descriptive)
```

```python
# Step 3: Detect outliers
lesser, greater = Univariate.outliercolnames(descriptive, quan)
print("Lesser outliers:", lesser)
print("Greater outliers:", greater)
```

```python
# Step 4: Cap outliers
data_capped = Univariate.cap_outliers(dataset, descriptive, lesser, greater)
print(data_capped)
```

```python
# create freqTable
freqTable=pd.DataFrame(columns=["Unique_Values","Frequecy","Relative_Frequeny","CumSum"])
freqTable
```

```python
#Find "Unique_Values","Frequecy","Relative_Frequeny","CumSum" for the table freqTable
freqTable["Unique_Values"]=dataset["ssc_p"].value_counts().index
freqTable["Frequecy"]=dataset["ssc_p"].value_counts().values
freqTable["Relative_Frequeny"]=(freqTable["Frequecy"]/103)
freqTable["CumSum"]=freqTable["Relative_Frequeny"].cumsum()
```
